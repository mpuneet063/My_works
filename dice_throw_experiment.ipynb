```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##conducting die throw experiment

#outcomes of two dices
exp = []
res_sum = []

#defining dice throw experiment
def expr(x):
    for n in range(1, x):
        x = np.random.randint(1, 7)  # Dice 1 (1 to 6 inclusive)
        y = np.random.randint(1, 7)  # Dice 2 (1 to 6 inclusive)
        total = x + y
        res_sum.append(total)
        exp.append(n)
```

```python
expr(101)
data = {'x':exp, 'y':res_sum}
df = pd.DataFrame(data)
print(df.columns)
#count occurence of each (x,y) pair
counts = df.groupby(['x', 'y']).size().reset_index(name='count')

#plotting
fig, axes = plt.subplots(1,2,figsize=(15,6))

#plotting the sum of outcomes
xticks = [i*10 for i in range(1,11)]
yticks = [j for j in range(2,13)]

axes[0].scatter(counts['x'], counts['y'], alpha=0.6, edgecolors='blue')
axes[0].set_xticks(xticks)
axes[0].set_yticks(yticks)
axes[0].set_title('Sum of outcomes on two dice vs number of experiments',{'fontsize':'18',
                                                 'font':'Times New Roman',
                                                 'fontweight':'25',
                                                 'color':'black'})
axes[0].set_xlabel('number of experiments')
axes[0].set_ylabel('sum of outcomes')

#plotting the frequency of sums
z = pd.Series(res_sum).value_counts().sort_index()
x_tick = [i for i in range(2,13)]
y_tick = []
axes[1].scatter(z.index,z.values,s=z.values*100,alpha=0.3,edgecolors='k')
axes[1].set_xticks(x_tick)
axes[1].set_yticks(y_tick)
axes[1].set_title('Frequency of a Sum',{'fontsize':'18',
                                                 'font':'Times New Roman',
                                                 'fontweight':'25',
                                                 'color':'black'})
axes[1].set_xlabel('sum of outcomes')
axes[1].set_ylabel('frequency of the sum')
```

### Observation
##### By conducting the dice throw experiment multiple times, we can observe that as the number of experiment increases, the frequency to obtain a certain sum of outcomes of the two dice starts to follow a pattern which resembles the trajectory of a projectile in air, i.e. the frequency increases till it obtains a maximum (namely a sum of 7) and then it decreases back to the minimum of 2. The result might lead to an interesting conclusion that though an experiment as random as a die throw can lead to a deterministic result which should make us wonder, is the world that we live in a deterministic world or non-deterministic?
